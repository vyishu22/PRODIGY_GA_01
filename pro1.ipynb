```python
import torch
from transformers import GPT2Tokenizer, GPT2LMHeadModel
```

```python
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")
```

```python
text = """
Artificial intelligence is one of the most important technologies of the modern world.
It enables machines to perform tasks that normally require human intelligence such as
learning, reasoning, problem solving, and decision making. AI systems are built using
large amounts of data and advanced algorithms that allow computers to recognize
patterns and improve their performance over time.

Machine learning is a major branch of artificial intelligence. It focuses on creating
models that can learn from data without being explicitly programmed. These models are
used in many applications such as recommendation systems, spam detection, speech
recognition, and image classification. Deep learning is a specialized area of machine
learning that uses neural networks with many layers to analyze complex data like
images, audio, and text.

Today, artificial intelligence is widely used in different industries. In healthcare,
AI helps doctors detect diseases earlier and analyze medical images more accurately.
In finance, AI systems help detect fraud and automate trading decisions. In education,
AI-powered platforms provide personalized learning experiences for students.

As technology continues to grow, artificial intelligence will play an even bigger role
in improving productivity, solving global problems, and transforming the way people
live and work in the future.
"""
```

```python
inputs = tokenizer(text, return_tensors="pt")
```

```python
optimizer = torch.optim.Adam(model.parameters(), lr=5e-5)

model.train()

for epoch in range(50):

    outputs = model(**inputs, labels=inputs["input_ids"])
    loss = outputs.loss

    loss.backward()
    optimizer.step()
    optimizer.zero_grad()

    print(f"Epoch {epoch} Loss: {loss.item()}")
```

```python
model.eval()

prompt = "Machine learning"

input_ids = tokenizer.encode(prompt, return_tensors="pt")

output = model.generate(
    input_ids,
    max_length=50,
    num_return_sequences=1,
    no_repeat_ngram_size=2
)

print(tokenizer.decode(output[0]))
```
